## <font color = 'green'>World Cup Basic Queries</font>

*1)  What player on a team with “ia” in the team name played less than 200 minutes and made more than 100 passes? Return the player surname. Note: To check if attribute A contains string S use "A like '%S%'"*

```python
%%sql
select surname from playersExt where team like '%ia%' and minutes < 200 and passes > 100
```

*2) Find all players who made more than 10 shots. Return all information about the players in
descending order of shots made.*

```python
%%sql
select * from PlayersExt where shots > 10 order by shots desc
```

*3) Find the goalkeepers of teams that played more than four games. List the surname of the goalkeeper, the team, and the number of minutes the goalkeeper played.*

```python
%%sql
select surname, team, minutes from PlayersExt where position = 'goalkeeper' and games > 4
```

*4) How many players who play on a team with ranking <10 played more than 350 minutes? Return one number in a column named 'superstar'.*

```python
%%sql
select count() as superstar from PlayersExt where ranking < 10 and minutes > 350
```

*5) What is the average number of passes made by forwards? By midfielders? Write one query that gives both values with the corresponding position.*

```python
%%sql
select position, avg(passes) from PlayersExt
where position = "forward" or position = "midfielder" group by position
```

*6) Which team has the highest average number of passes per minute played? Return the team and average passes per minute. Hints: (1) You can compute a team's average number of passes per minute played by dividing the total number of passes by the total number of minutes. To force floating point division, multiply one operand by 1.0. (2) Consider using Limit.*

```python
%%sql
select team, 1.0*sum(passes)/sum(minutes)
from PlayersExt
group by team
order by 1.0*sum(passes)/sum(minutes) desc
limit 1
```

## <font color = 'green'>World Cup Advanced Queries</font>

*1) Find all pairs of teams who have the same number of goalsFor as
each other and the same number of goalsAgainst as each other.
Return the teams and numbers of goalsFor and goalsAgainst.
Make sure to return each pair only once.*

```python
%%sql
select T1.team, T2.team, T1.goalsFor, T1.goalsAgainst from playersExt T1, playersExt T2 
where T1.goalsFor = T2.goalsFor and T1.goalsAgainst = T2.goalsAgainst and T1.team < T2.team
```

*2) Find all teams with ranking <30 where no player made more than 150 passes. Return the team and ranking.*

```python
%%sql
select team, ranking from playersExt T where ranking < 30 and
not exists (select * from PlayersExt P where P.team = T.team and P.passes > 150)
```

*3) Which team has the highest ratio of goalsFor to goalsAgainst?*

```python
%%sql
select team from playersExt
where goalsFor/goalsAgainst = (select max(goalsFor/goalsAgainst) from playersExt)
```

*4) Find all team-position pairs where the average number of passes made by players in that position on that team is greater than 150. Return the team-position pairs.*

```python
%%sql
select team, position from PlayersExt group by team, position having avg(passes) > 150
```

*5) Find all teams whose defenders averaged more than 150 passes. Return the team and average number of passes by defenders, in descending order of average passes.*

```python
%%sql
select team, avg(passes) from PlayersExt where position = "defender" group by team
having avg(passes) > 150 order by avg(passes) desc
```

## <font color = 'green'>Titanic Advanced Queries</font>

*1) Are there any pairs of passengers with the same last name where one is in first class and the other is in third class? If so, return the last name and the two first names. Label the first-name columns 'first' for the passenger in first class and 'third' for the passenger in third class.*

```python
%%sql
select T1.last, T1.first, T2.first as 'third' from Titanic T1, Titanic T2
where T1.last = T2.last and T1.class = 1 and T2.class = 3
```

*2) Which embarkation cities have more than 40 passengers whose age is missing? Reminder: Blanks in SQL tables are given a special value called 'null', and conditions 'A is null' and 'A is not null' can be used in Where clauses to check whether attribute A has the 'null' value.*

```python
%%sql
select distinct embarked from Titanic T1
where 40 < (select count() from Titanic T2 where T2.embarked = T1.embarked and age is null)
```

*3) Find all classes where the average fare paid by passengers in that class was more than twice the overall average or less than half the overall average.*

```python
%%sql
select class from Titanic group by class
having avg(fare)>2*(select avg(fare) from Titanic) or avg(fare)<0.5*(select avg(fare) from Titanic)
```

*4) EXTRA CREDIT CHALLENGE: List each class and its survival rate, i.e., the fraction of passengers in that class who survived.*

```python
%%sql
select yes.class, 1.0*Scount / Tcount as survival
from
(select class, count() as Scount
from Titanic
where survived = 'yes'
group by class)yes,
(select class, count() as Tcount
from Titanic
group by class)total
where yes.class = total.class
```
