### Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Import Dataset

```python
df_team = pd.read_csv('Teams.csv')
df_player = pd.read_csv('Players.csv')
```

### Correlations (Teams.csv)

```python
df_team.head(5)
```

```python
#correlation in team dataset
corr_team = df_team.corr()
corr_team
```

### Correlation Heatmap

```python
plt.figure(figsize = (10,8))
ax = sns.heatmap(corr_team, annot = True, fmt = '.0%')
```

### Seaborn Pairplot

```python
#compare pairplot and heatmap positive and negative correlation
sns.pairplot (corr_team, height = 1.3)
```

### Correlations (Players.csv)

### Correlation Heatmap
